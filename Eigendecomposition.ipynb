# Code Challenge 1 - Eigenvalues of diagonal and triangular matrices 
### Shows that in diagonal and triangular matrices, the eigenvalues are the numbers located on the diagonal 
#### Note: The order of the eigenvalues printed out may be different than their occurence in the matrices

```python
# Genrate a diagonal matrix (2x2) 
diag1 = np.diag([2,1])
print(diag1,'\n')

# Compute its eigenvalues 
diag1_eigvals = np.linalg.eig(diag1)
print('Eigenvalues \n',diag1_eigvals[0])
```

```python
# Expand to an NxN diagnonal matrix 
diag2 = np.diag(np.arange(2,24))
print(diag2,'\n') 

# Compute its eigenvalues 
diag2_eigvals = np.linalg.eig(diag2)
print('Eigenvalues \n',diag2_eigvals[0])
```

```python
# Generate a lower triangular matrix 
triag1 = np.tril(np.random.randn(5,5))
print(triag1,'\n')

# Compute its eigenvalues 
triag1_eigvals = np.linalg.eig(triag1)
print('Eigenvalues \n',triag1_eigvals[0])
```

```python
# Generate an upper triangular matrix 
triag2 = np.triu(np.random.randn(5,5))
print(triag2,'\n') 

# Compute its eigenvalues 
triag2_eigvals = np.linalg.eig(triag2)
print('Eigenvalues \n',triag2_eigvals[0])
```

# Code Challenge 2 - Eigenvalues of random matrices
### Will plot a series of complex eigenvalues. They come in conjugate pairs, i.e., for every value, there is a corresponding value that has the same real part but has the sign flipped for the imaginary part. Thus, the circle is fully symmetrical, i.e., above the zero line is a mirror of below the zero line. 

```python
# Generate a 40x40 random matrices, extract their eigenvalues, make a plot
#Repeat this lots of times for more random matrices, plot them in the same plot 

n = 40 

for i in range(200):
    A = np.random.randn(n,n)
    eigvals = np.linalg.eig(A)
    eigvals = eigvals[0]
    plt.plot(np.real(eigvals),np.imag(eigvals),'s',markersize=2)
    
plt.axis('square')
plt.show()
```

# Code Challenge 3 - Show the process of Eigendecomposition (a.k.a Diagonalisation)

```python
"""
Code Challenge 3 - Show the process of eigendecomposition. 
You start with a square Matrix (A) and you apply a set of rotation that are contained 
in the V matrix of eigenvectors. Once A has been moved to the space defined by the column's of 
matrix V, A looks like a diagonal matrix. You then need V inverse to get out of the space and back to
the space where A is. 

I am making matrix A a symmetric matrix as it will have real values in the output. 
This will make visualisation easier. 
"""
# Create a random matrix A 
A = np.round(10*np.random.randn(4,4))
A = A.T@A

#Get the eigenvectors and eigenvalues 
lambdas, V = np.linalg.eig(A) 

#Turn the singular array of lambdas into a diagonal matrix 
Lambdas = np.diag(lambdas) 

# Get the inverse of V 
Vinv = np.linalg.inv(V) 

# Test the reconstruction of A through making a new matrix from the formula V*Lambda*V^-1
Arec = V@Lambdas@Vinv
```

```python
# Plotting the above to show the results 

# Plotting A 
plt.subplot(1,2,1,) 
plt.imshow(A) 
plt.axis('off')
plt.title('A') 

# Plotting the reconstruction 
plt.subplot(1,2,2) 
plt.imshow(Arec)
plt.axis('off')
plt.title('$V\Lambda V^{-1}$')

plt.show()

print("""
Plotting A and Arec(The reconstructed product of the eigendecomposition of A) will return the same matrices. 
This shows that Eigendecomposition is the process of factorising a matrix to be represented in 
terms of its eigenvalues and eigenvectors. 
""")
```

# Code Challenge 4 - Symmetric matrices
#### A sample of the code required to obtain a symmetric matrix. It shows that the matrix of eigenvectors in a symmetric matrix (V) will be a symmetric matrix 

```python
#Create a random matrix 
A = np.random.randn(22,22) 

# Make a symmetric matrix using the additive method 
Asym = A+A.T

# Eigendecomposition for the symmetric matrix 
lambdas1, V1 = np.linalg.eig(Asym) 

#Eigendecomposition for the non symmetric matrix 
lambdas2, V2 = np.linalg.eig(A) 
```

```python
# Show the matrices as an image 

# Symmetric matrix 
print("FOR A SYMMETRIC MATRIX")
# Original matrix (Asym) 
plt.subplot(1,3,1)
plt.imshow(Asym) 
plt.axis('off')
plt.title('A (Symmetric)')

# Eigenvectors Matrix (V1) 
plt.subplot(1,3,2)
plt.imshow(V1) 
plt.axis('off') 
plt.title('V') 

# Multiplying the eigenvectors matrix (V1) by its own transpose 
plt.subplot(1,3,3)
plt.imshow(V1@V1.T)
plt.axis('off')
plt.title('$VV^T$')

plt.show()

print("""
The eigenvectors matrix times its own Transpose will give the Identity matrix. This shows that the 
transpose of the eigenvectors matrix is the same as its inverse. This is only the case because the original 
matrix A is a symmetric matrix. See sample of a non-symmetric matrix below.\n 
""")

# Non-symmetric matrix 
print("FOR A NON-SYMMETRIC MATRIX")

#Original matrix (A) 
plt.subplot(1,3,1) 
plt.imshow(A)
plt.axis('off')
plt.title('A (Non-Symmetric)')

# Eigenvectors Matrix (V2) 
plt.subplot(1,3,2)
plt.imshow(V2.real)
plt.axis('off')
plt.title('V')

# Multiplying the eigenvectors matrix (V2) by its own transpose 
plt.subplot(1,3,3)
plt.imshow(V2.real@V2.real.T)
plt.axis('off')
plt.title('$VV^T$')

plt.show()
```

# Code Challenge 5 - Trace and determinant, eigenvalues sum and product 
#### Shows trace(A) is the same thing as the sum of the eigenvalues. 
#### Also shows that determintant(A) is the same thing as the product of all the eigenvalues. 

```python
# Generate a square matrix A
m = 5
A = np.random.randn(m,m)

#Eigendecomposition (Only need the eigenvalues for this exercise) 
evals, evecs = np.linalg.eig(A) 

# Show trace(A) == sum(evals)
sum_evals = sum(evals)
trace_evals = np.matrix.trace(A)
print(sum_evals,trace_evals,'\n')

#Show the determinant(A) == product(evals) 
prod_evals = np.prod(evals)
det_evals = np.linalg.det(A)
print(prod_evals,det_evals)
```

# Code Challenge 6 - Generalized Eigendecomposition in small and large matrices
#### Extract the eigenvalues and eigenvectors from the following two equations and determine whether they are the same: $Sv=\lambda Rv$ and $R^-1Sv=\lambda v$
#### Repeat this process twice. Once for a very small matrix (2x2) and also for a larger matrix. 

```python
"""
First, do it for the small matrices. 
"""
# Define matrices S and R 
S = np.random.randn(2,2) 
R = np.random.randn(2,2) 

# GED 
evals1, evecs1 = scipy.linalg.eig(S,R) 
evals2, evecs2 = scipy.linalg.eig(np.linalg.inv(R)@S) 

# Print eigenvalues for both (Should be identical) 
print("Eigenvalues for both equations")
print(evals1,'\n')
print(evals2,'\n')
print("Eigenvalues are the same")
```
