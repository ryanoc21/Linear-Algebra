```python
# Compare Lambda and Sigma 
print("Comparing the Eigenvalues and the Singular Values")
print("Eigenvalues:\n",np.diag(lambdas)) 
print("\n Singular Values:\n",np.diag(Sigma))
print("\n They are the same values along the diagonal but they are in different order")
```

```python
# Compare left and right Singular Values (U and VT )
print("Comparing the left and right Singular Values")
print("Left Singular Values (U):\n",U)
print("\nRight Singular Values (V^T):\n",VT)
print("\n",np.round(U-VT.T)) # You need to transpose it back for the comparison to work
```

```python
# Compare Left Singular Values and Eigenvectors. 

# You will need to sort the eigenvectors to make this comparison 
sidx = np.argsort(lambdas)[::-1]
lambdas = lambdas[sidx]
eigvecs = eigvecs[:,sidx]


print("Comparing the left Singular Values and the Eigenvectors")
print("Left Singular Values (U):\n",U)
print("\nEigenvectors (V):",eigvecs)
print("\n",np.round(U-eigvecs))

print("""
The discrepancy in the third column is due to the difference in the sign,i.e., in U they are positive and V they 
are negative. If you change the sign in the subtraction you will get the third column to be 0. This does not change 
the fact that the Left Singular Values and the Eigenvectors are the same for square symmetric matrices
""")
```

#### The above shows that for a square symmetric matrix, eigendecomposition is almost the same thing as SVD with two implementational differences. 
#### 1. There are discrepancies possibile sorting of singular values and eigenvalues 
#### 2. There can be discrepancies in the sign of individual values contained in U and the eigenvectors matrix which may slightly distort the interpretation, however, it is easy to see that both outputs are similar when using a relatively small matrix, regardless of the sign. 
This supports the notion that SVD is essentially a version of Eigendecomposition that works for rectangular matrices

# 3 - Relation between singular values and eigenvalues
#### Expands on the relationships established in the previous code challenge. Ultimately it shows that for square, symmetric matrices, the eigenvalues and the singular values are the same when the eigenvalues are calculated from the transpose of the original matrix A

```python
# case 1: eig(ATA) vs svd(A)
A = np.random.randn(2,3) 

print(np.sort(np.linalg.eig(A.T@A)[0]))
print(np.sort(np.linalg.svd(A)[1])**2)

print("The eigenvalues of ATA are the same as the singular values of A (when they are squared)")
```

```python
# case 2: eig(ATA) vs svd(ATA)

print(np.sort(np.linalg.eig(A.T@A)[0]))
print(np.sort(np.linalg.svd(A.T@A)[1]))

print("""
These are the same values without having to square the singular values. 
""")
```

The above code snippets show that the eigenvalues are singular values are the same (albeit with trivial differences, i.e., having to square the singular values in case 1). However, this is only the case with ATA (A transpose times A). The following code will show situtations where they are different.

```python
# Case 3a: eig(A) vs svd(A) with real valued eigenvalues 

# Need a square matrix for eigendecomposition 
A = np.random.randn(3,3) 

# Make symmetric (ensures only real values output) 
A = A+A.T

print(np.sort(np.linalg.eig(A)[0]))
print(np.sort(np.linalg.svd(A)[1]))
```

```python
# Case 3b: eig(A) vs svd(A) with complex valued eigenvalues 

# Need a square matrix for eigendecomposition 
A = np.random.randn(3,3) 

print(np.sort(np.linalg.eig(A)[0]))
print(np.sort(np.linalg.svd(A)[1]))
```

# 4 - U from Eigendecomposition of ATA

```python
# Create a 3x6 matrix A
m = 3
n = 6 
A = np.random.randn(m,n) 

# Perform full SVD (Variables: Us,Ss,Vs) 
Us, Ss, Vs = np.linalg.svd(A) 
 

# Perform eig of ATA (Variables: V,L)
Lambdas, Eigvecs = np.linalg.eig(A.T@A) 


# You will need to sort the eigenvectors to make this comparison 
sidx = np.argsort(Lambdas)[::-1]
Lambdas = Lambdas[sidx]
Eigvecs = Eigvecs[:,sidx]
```

```python
# Confirm that Vs = V
print(np.round(Vs.T-Eigvecs,2))
```

```python
# Check the relationship between Singular values and Eigenvalues 
print(Ss**2) 
print('\n',Lambdas)

print("""
These are the same, just note that the eigenvalues have a complex zero displayed, the other three disimilar values
are 0 plus computer rounding error. 
""")
```

```python
# Create U using only A, V and L 
U = np.zeros((m,m))

for i in range(m): 
    U[:,i] = A@Eigvecs[:,i].T/np.sqrt(Lambdas[i])
```

```python
# Confirm that U == Us 
print(np.round(U-Us,2))
```

# 5 - (a) Spectral Theory of Matrices and (b) Low Rank Approximations
#### Essentialy the idea that matrices are layers of representations as the sum of many matrices, all with rank 1. It is intuitively analagous to the layer perspective of matrix multiplication which represents matrices as the product of the information contained within the matrices that are multiplied together. 
#### SVD will tell you important directions in the matrix that are combinations across columns or rows, ie., $U$ and$ V^T$. The importance of these directions is denoted by the size of their corresponding singular values (contained in $\Sigma$), i.e., the larger the singular values, the more important the direction is to the overall information contained within the matrix and vice-versa for smaller singular values

## (a) Spectral Theory of Matrices
**The code for this sections was taken from the class notes**

```python
### NOTE - THIS CODE WAS COPIED FROM THE CLASS NOTES *** 

# matrix sizes
m = 40
n = 30

# define a 2D Gaussian for smoothing
k = int( (m+n)/4 )
xx = np.linspace(-3,3,k)
[X,Y] = np.meshgrid(xx,xx)
g2d = np.exp(-(X**2 + Y**2) / (k/8))

# matrix
A = scipy.signal.convolve2d(np.random.randn(m,n),g2d,'same')

plt.imshow(A)
plt.show()
```

```python
### NOTE - THIS CODE WAS ALSO COPIED FROM THE CLASS NOTES *** 

# SVD (remember that V comes out transposed)
U,S,V = np.linalg.svd(A)


# show the constituent matrices
fig = plt.subplots(1,figsize=(10,6))

plt.subplot(241)
plt.imshow(A)
plt.title('A')

plt.subplot(242)
plt.imshow(U)
plt.title('U')

plt.subplot(243)
plt.imshow(np.diag(S))
plt.title('$\Sigma$')

plt.subplot(244)
plt.imshow(V)
plt.title('V$^T$')

plt.subplot(212)
plt.plot(S,'ks-')
plt.xlabel('Component number')
plt.ylabel('$\sigma$')
plt.title('"Scree plot" of singular values')

plt.show()
```

```python
### NOTE - THIS CODE WAS ALSO COPIED FROM THE CLASS NOTES *** 

rank1mats = np.zeros((5,m,n))

for i in range(0,5):
    
    # create rank1 matrix
    rank1mats[i,:,:] = np.outer( U[:,i]*S[i],V[i,:] )
    
    plt.subplot2grid((2,5),(0,i))
    plt.imshow(rank1mats[i,:,:],vmin=-5,vmax=5)
    plt.axis('off')
    plt.title('C.%d' %(i+1))
    
    plt.subplot2grid((2,5),(1,i))
    imdat = np.sum(rank1mats[:i+1,:,:],axis=0)
    plt.imshow(imdat,vmin=-5,vmax=5)
    plt.axis('off')
    plt.title('Cs 1:%d' %(i+1))

plt.show()

# then show the original
plt.imshow(A,vmin=-5,vmax=5)
plt.show()
```

## (b) Low Rank Approximations 
**This code is my own**

```python
# Make a large matrix 
m = 40
n = 30 
A = np.random.randn(m,n) 

# Perform SVD 
U,S,V = np.linalg.svd(A) 

# Plot 
plt.subplot(1,4,1) 
plt.imshow(A) 
plt.title("A") 
plt.axis("off")

plt.subplot(1,4,2) 
plt.imshow(U) 
plt.title("U")
plt.axis("off")

plt.subplot(1,4,3) 
plt.imshow(np.diag(S)) 
plt.title("$\Sigma$") 
plt.axis("off")

plt.subplot(1,4,4) 
plt.imshow(V) 
plt.title("$V^T$") 
plt.axis("off")

plt.show()
```

```python
# Performing a low rank approximation 

# Number of components maintained in the matrices 
no_Comps = 5 

# Create reduced rank matrices 
Ur = U[:,0:no_Comps]
Sr = S[0:no_Comps]
Vr = V[0:no_Comps,:]
```

```python
# Make a low rank approximation of matrix A 
Ar = Ur@np.diag(Sr)@Vr
```

```python
# Show reduced rank as images 
print("\nReduced Rank Approximation of Matrix A")
plt.subplot(1,4,1) 
plt.imshow(Ar) 
plt.title("A") 
plt.axis("off")

plt.subplot(1,4,2) 
plt.imshow(Ur) 
plt.title("U")
plt.axis("off")

plt.subplot(1,4,3) 
plt.imshow(np.diag(Sr))
plt.title("$\Sigma$") 
plt.axis("off")

plt.subplot(1,4,4) 
plt.imshow(Vr) 
plt.title("$V^T$") 
plt.axis("off")

plt.show()
```

```python
# Get an error matrix, ie, the difference between the full and reduced rank approximation 
errormap = (Ar-A)**2
```

```python
# Plot full rank, reduced rank approximation and error matrix 
plt.subplot(1,3,1) 
plt.imshow(A) 
plt.title(f"Rank = {np.linalg.matrix_rank(A)}") 
plt.axis("off")

plt.subplot(1,3,2) 
plt.imshow(Ar) 
plt.title(f"Reduced Rank = {np.linalg.matrix_rank(Ar)}") 
plt.axis("off")

plt.subplot(1,3,3) 
plt.imshow(errormap) 
plt.title("Error")
plt.axis("off")

plt.show()
```

# 6 - Converting Singular Values to Percent Variance 
#### Using the method of summing all singular values, dividing each singular value by the sum and multiplying by 100. 
#### Formula: $\sigma i = 100\sigma i/\Sigma \sigma i$

```python
# Create a matrix 
m = 40
n = 30
A = np.random.randn(m,n) 

# Scale the matrix 
Ascaled = A*100  # You can change this value to show how the percent of variance will not change. 

# Perford SVD 
U,S,V = np.linalg.svd(Ascaled) 
```

```python
# Convert to percent Variance 
Svar = 100*S/np.sum(S) 

# Plot the raw singular values and the percent variance singular values to compare 
fig = plt.subplots(1,figsize=(7,5))

plt.subplot(2,1,1) 
plt.plot(S,'ks-')
plt.xlabel("Componenet Number")
plt.ylabel("$\sigma$")
plt.title("Raw Singular Values")

plt.subplot(2,1,2) 
plt.plot(Svar,'ks-')
plt.xlabel("Componenet Number")
plt.ylabel("$\sigma$ (% of total)")
plt.title("Percent Variance Normalised Singular Values")

plt.tight_layout()
```

# 7 - Determine the conditions when the right singular vectors ($U$) and the left singular vectors ($V^T$) can be multiplied. 
#### - Examine what the resulting matrix looks like.
#### - What is its norm? 
#### - Is it orthogonal?

```python
# Generate a matrix such that U*VT is valid (i.e., a square matrix) 
m = 5
A = np.random.randn(m,m) 

# Perform SVD 
U,S,VT = np.linalg.svd(A) 

# Compute the norm of U, the norm of V and the norm of U*VT 
print(f"Norm of U: {np.linalg.norm(U,2)}")
print(f"Norm of V^T: {np.linalg.norm(VT,2)}")
print(f"Norm of U*VT: {np.linalg.norm(U@VT,2)}")
```

```python
# Test for the result of the matrix U*UT, V*VT, U*VT 
UUT = U@U.T
VVT = VT.T@VT
UVT = U@VT 

print(np.round(UUT,2),'\n')
print(np.round(VVT,2),'\n')
print(np.round(UVT,2),'\n')
print("Making UVT symmetric will give you the identity matrix")
print(np.round(UVT@UVT.T,2),'\n')
```

```python
plt.subplot(1,4,1)
plt.imshow(UUT)
plt.title("U*UT")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(VVT)
plt.title("V*VT")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(UVT)
plt.title("U*VT")
plt.axis("off")

plt.subplot(1,4,4) 
plt.imshow(UVT@UVT.T) 
plt.title("(U*VT)*(U*VT)^T")
plt.axis("off")

plt.tight_layout()
plt.show()
```

U*UT and V*VT are orthogonal, thus multiplying them together will give the identity matrix. U*VT will not give you the identity matrix, however, making it symmetric will, as shown above

# 8 - Create a matrix with a desired condition number

```python
# Create a random matrix with a specified condition number 
m = 10
n = 23
cond_num = 42 # (The answer to life, the universe and everything.)

# Create singular vectors matrices 
U,junk = np.linalg.qr(np.random.randn(m,m))
V, junk = np.linalg.qr(np.random.randn(n,n))

# Create singular values matrix 
s = np.linspace(cond_num,1,np.minimum(m,n))
S = np.zeros((m,n)) 

for i in range(len(s)):
    S[i,i] = s[i]
    
# Create matrix A out of the generated SVD matrices
A = U@S@np.transpose(V) 
print("Condition Number of A: ",np.round(np.linalg.cond(A),2))


# Make images of the matrix itself and the 3 SVD matrices
plt.subplot(1,4,1) 
plt.imshow(A) 
plt.axis('off')
plt.title(f"A (condnum={np.round(np.linalg.cond(A),2)})")

plt.subplot(1,4,2) 
plt.imshow(U) 
plt.axis('off')
plt.title("U")

plt.subplot(1,4,3) 
plt.imshow(S) 
plt.axis('off')
plt.title("$\\Sigma$")

plt.subplot(1,4,4) 
plt.imshow(np.transpose(V)) 
plt.axis('off')
plt.title("$V^T$")

plt.tight_layout()
plt.show()
```
